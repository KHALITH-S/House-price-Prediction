```python
remove=['date','yr_renovated','street','city','statezip','country']
df=df.drop(remove,axis=1)
```

```python
import scipy.stats as stats
z_score = stats.zscore(df)
print("before :",df.shape)
outlier=df[(z_score>3).any(axis=1)]
df=df[(z_score<=3).all(axis=1)]
print("after :",df.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

scaler.fit(df)
df_scaler=pd.DataFrame(scaler.transform(df),columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x=df.drop('price',axis=1)
y=df['price']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
print("train =",x_train.shape)
print("test =",x_test.shape)
```

```python
models=[
    ('Random Forest',RandomForestRegressor()),
    ('Linear Regression',LinearRegression()),
    ('Decision Tree',DecisionTreeRegressor()),
    ('KNN',KNeighborsRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
for name,model in models:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(name.upper())
    print("Mean sqa:",mean_squared_error(y_test,y_pred))
    print("Mean abb:",mean_absolute_error(y_test,y_pred))
    print("R-squa:",r2_score(y_test,y_pred))

    print('\n')

```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(y_test, label="Actual Price", color="blue", fill=True, alpha=0.4)
sns.kdeplot(y_pred, label="Predicted Price", color="green", fill=True, alpha=0.4)
plt.xlabel("Price")
plt.ylabel("Density")
plt.title("Distribution of Actual vs Predicted Prices")
plt.legend()
plt.show()
```
